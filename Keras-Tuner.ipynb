# Neural Network

#### Model Training 
 * Numbber of layers
     * A neural network consists of:
        * Input Layer
        * Hidden Layers
        * Output Layer
 * Loss Function
   * Measures how far predictions are from actual values
   * Model tries to minimize loss
   * Ex:
     * Binary Classification -- binary_crossentropy
     * Multi-class Classification -- categorical_crossentropy
     * Regression -- mean_squared_error (MSE)
 * Activation
    * Introduce non-linearity into the model
 * Optimizer
   * Updates weights to reduce loss

## Keras Tuning 
### What is keras Tuning? 
* Keras Tuning Used to find the best hyperparameters for a model
* It Improves model performance automatically
* Model tries to : Find Optimal architecture, Reduce Validation loss,Increase accuracy

### What Can Be Tuned
* Number of layers
* Number of neurons
* Learning rate
* Batch size
* Dropout rate
* Activation functions

## Choosing Best Optimizer

```python
def build_model(hp):
  model = keras.Sequential([
    keras.Input(shape= (8, )),
    keras.layers.Dense(32, activation="relu", name= "H1"),
    keras.layers.Dense(1, activation="sigmoid", name = "Output")])

  Loss = "binary_crossentropy"
  Optimizer = hp.Choice("optimizer", ["adam", "sgd", "rmsprop", "adadelta"])
  Metric = ["accuracy"]

  model.compile(loss = Loss, optimizer= Optimizer, metrics = Metric)

  return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective = "val_accuracy",
                        max_trials=5)
tuner
```

```python
tuner.search(X_train,
             y_train,
             epochs = 5,
             validation_data= (X_test, y_test))
```

```python
model = tuner.get_best_models()[0]
model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs = 10,
    initial_epoch = 5,
    validation_data = (X_test, y_test),
    batch_size = 32
)
```

## Choosing Number of Layers/Neurons

```python
def build_model(hp):

  Units = hp.Int("units", min_value = 8, max_value = 128, step = 8)
  model = keras.Sequential([
    keras.Input(shape= (8, )),
    keras.layers.Dense(units=Units, activation="relu", name= "H1"),
    keras.layers.Dense(1, activation="sigmoid", name = "Output")])

  Loss = "binary_crossentropy"
  # Optimizer = hp.Choice("optimizer", ["adam", "sgd", "rmsprop", "adadelta"])
  Optimizer = "sgd"
  Metric = ["accuracy"]

  model.compile(loss = Loss, optimizer= Optimizer, metrics = Metric)

  return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective = "val_accuracy",
                        max_trials= 5,
                        directory = "mydir")
```

```python
tuner.search(X_train,
             y_train,
             epochs = 5,
             validation_data = (X_test, y_test)
             )
```

```python
model = tuner.get_best_models()[0]
model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs = 10,
    initial_epoch = 5,
    validation_data = (X_test, y_test),
    batch_size = 32
)
```
